```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv("AB_NYC_2019.csv");
data.head()
```

```python
# Making the titles of the columns uppercase 
data.columns = data.columns.str.upper()
```

```python
# Removing duplicates from the dataset
data.duplicated().sum()

# No duplicates
```

```python
# Removing NaN values from the dataset
data.isna().sum()

data.dropna(inplace = True) # Dropping the NaN values
```

```python
data.columns
```

<h1> Price Distribution of Listings </h1>

<h3> The vast majority of properties is concentrated in the lower price range. This means most listings on the platform are relatively affordable catering for budget conscious customers </h3>

```python
import matplotlib.pyplot as plt

plt.figure()
plt.hist(data['PRICE'], bins=50)

plt.title('Price Distribution')
plt.xlabel('PRICE')
plt.ylabel('FREQUENCY')

plt.show()
```

<h1> Room Types </h1>

<h3> The majority of properties acquired by customers are Entire homes while a little fancy shared properties </h3>

```python
room_counts = data['ROOM_TYPE'].value_counts()

plt.figure()
plt.bar(room_counts.index, room_counts.values)

plt.title('Number of Listings by Room Type')
plt.xlabel('ROOM_TYPE')
plt.ylabel('COUNT')

plt.xticks(rotation=30)
plt.show()
```

<h1> Price VS Number of views </h1>

<h3> Listings with lower prices had ore reviews. This is also because most customers acquire relatively affordable properties </h3>

```python
plt.figure()
plt.scatter(data['NUMBER_OF_REVIEWS'], data['PRICE'])

plt.title('Price vs Number of Reviews')
plt.xlabel('NUMBER_OF_REVIEWS')
plt.ylabel('PRICE')

plt.show()
```
