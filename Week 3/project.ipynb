```python
import tkinter as tk
import pandas as pd

# Load the spreadsheet
file_path = r"C:\Users\NEW\Documents\m.babajideCOS102\Week 3\Project\Users.csv"
data = pd.read_csv(file_path)

# Function to check user
def check_user():
    username = username_entry.get()
    department = department_entry.get()

    # Search the spreadsheet
    for index, row in data.iterrows():
        if row["Username"] == username and row["Department"] == department:
            result_label.config(text="Welcome " + row["Name"])
            found = True
            break

    if found == False:
        result_label.config(text="User not found")

# Create window
window = tk.Tk()
window.title("User Login")

# Welcome text
welcome_label = tk.Label(window, text="Welcome! Please login")
welcome_label.pack()

# Username
username_label = tk.Label(window, text="Enter Username")
username_label.pack()

username_entry = tk.Entry(window)
username_entry.pack()

# Department
department_label = tk.Label(window, text="Enter Department")
department_label.pack()

department_entry = tk.Entry(window)
department_entry.pack()

# Login Button
login_button = tk.Button(window, text="Login", command=check_user)
login_button.pack()

# Result message
result_label = tk.Label(window, text="")
result_label.pack()

# Run program
window.mainloop()

    
```
