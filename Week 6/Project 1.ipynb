```python
import tkinter as tk
from tkinter import messagebox as msg
import random

class Employee:
    def __init__(self, input_name):
        self.user = input_name
        # Database of staff and duties
        self.staff_list = [
            "Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali", 
            "Andrew Ugwu", "Stella Mankinde", "Jane Akibo", "Ago James",
            "Michell Taiwo", "Abraham Jones", "Nicole Anide", "Kosi Korso", 
            "Adele Martins", "Emmanuel Ojo", "Ajayi Fatima"
        ]
        self.duty_list = ["Loading", "Transporting", "Reviewing Orders", "Customer Service", "Delivering Items"]

    def is_valid(self):
        """Checks if the entered name exists in the database."""
        return self.user in self.staff_list

    def log_presence(self):
        """Returns the attendance confirmation string."""
        return f"Status: {self.user} is marked as PRESENT for today."

    def get_random_task(self):
        """Selects a random task using an index as requested in the hint."""
        idx = random.randint(0, len(self.duty_list) - 1)
        return f"Assigned Duty: {self.duty_list[idx]}"

    def access_denied_msg(self):
        """Returns the refusal message."""
        return "Unauthorized: Name not found in the employee directory."

def process_login():
    # Retrieve name and format it to Title Case
    entered_name = name_box.get().strip().title()
    
    # Create the employee object
    worker = Employee(entered_name)

    if worker.is_valid():
        attendance_status = worker.log_presence()
        daily_task = worker.get_random_task()
        
        # Display success message
        msg.showinfo("Verification Successful", 
                     f"Welcome to the System!\n\n{attendance_status}\n{daily_task}")
    else:
        # Display error message
        msg.showerror("Login Failed", worker.access_denied_msg())

# --- User Interface Setup ---
app = tk.Tk()
app.title("Staff Identification System")
app.geometry("300x150")

# Adding a bit of padding for a cleaner look
tk.Label(app, text="Please enter your full name:", font=("Arial", 10)).pack(pady=5)

name_box = tk.Entry(app, width=30)
name_box.pack(pady=5)

# Trigger button
submit_btn = tk.Button(app, text="Verify Identity", bg="lightgray", command=process_login)
submit_btn.pack(pady=15)

app.mainloop()
```
