```python
import tkinter as tk
from tkinter import messagebox as msg

class ShippingCalculator:
    def __init__(self, dest, mass):
        self.dest = dest.lower() # Normalize to lowercase for easier checking
        self.mass = mass

    def get_price(self):
        # Pricing logic for PAU
        if "pau" in self.dest or "pan-atlantic" in self.dest:
            return 2000 if self.mass >= 10 else 1500
        
        # Pricing logic for Epe
        elif "epe" in self.dest:
            return 5000 if self.mass >= 10 else 4000
        
        return None

def process_calculation():
    user_location = loc_field.get().strip()
    
    try:
        user_weight = float(weight_field.get())
        if user_weight < 0:
            msg.showwarning("Warning", "Weight must be a positive number.")
            return
    except ValueError:
        msg.showerror("Input Error", "Please enter a numeric value for weight.")
        return

    # Create instance and calculate
    calc = ShippingCalculator(user_location, user_weight)
    total_cost = calc.get_price()

    if total_cost:
        msg.showinfo("Quote Ready", 
                     f"Destination: {user_location.title()}\n"
                     f"Package Weight: {user_weight}kg\n\n"
                     f"Total Shipping Fee: ₦{total_cost:,}")
    else:
        msg.showerror("Area Not Covered", "We currently only deliver to PAU and Epe.")

# --- Graphical Interface ---
window = tk.Tk()
window.title("Jane's Delivery Service")
window.geometry("350x200")

# UI Layout
tk.Label(window, text="Delivery Destination (e.g. PAU or Epe):").pack(pady=5)
loc_field = tk.Entry(window)
loc_field.pack()

tk.Label(window, text="Weight of Package (kg):").pack(pady=5)
weight_field = tk.Entry(window)
weight_field.pack()

calc_btn = tk.Button(window, text="Get Delivery Quote", command=process_calculation, bg="lightblue")
calc_btn.pack(pady=20)

window.mainloop()
```
