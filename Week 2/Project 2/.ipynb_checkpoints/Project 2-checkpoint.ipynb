```python
import numpy as np

def find_poly_roots():
    print("")
    print("Enter coefficients from the highest degree down to the constant term.")
    print("For a quadratic (ax^2 + bx + c), enter a, b, c.")
    print("For a cubic (ax^3 + bx^2 + cx + d), enter a, b, c, d.")
    print("For a quartic (ax^4 + bx^3 + cx^2 + dx + e), enter a, b, c, d, e.")

    # Get user input and convert to a list of floats
    coefficients = list(map(float, input("Enter coefficients separated by spaces: ").split()))

    # Check if the list of coefficients is empty
    if not coefficients:
        print("Error: No coefficients entered.")
        return

    # Use numpy.roots to find roots
    roots = np.roots(coefficients)
    print(f"\nThe roots of the polynomial are:\n{roots}")
    
find_poly_roots()
```
