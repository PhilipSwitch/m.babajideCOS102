```python



print("Welcome to this Program")
print("This is a program to calculate Amount")
print("Select form the list of Options input:")

print("SI for SImple Interest")
print("CI for Compound Interest")
print("AP for Annuity Plan")

input1 = input("Enter your option: SI, CI or AP").upper()

if input1 == "SI":
    # Simple Interest 
    P = float(input("Enter Principal (P):"))
    R = float(input("Enter Rate (R in %):"))
    T = float(input("Enter Time (T in years):"))

    A = P * (1 + (R / 100) * T)
    print(f"Amount (A) = {A}")

elif input1 == "CI":
    # Compound Interest
    P = float(input("Enter Principal (P): "))
    R = float(input("Enter Rate (R in %): "))  
    n = float(input("Enter number of compounding periods per year (n): "))
    t = float(input("Enter time in years (t): "))
    A = P * (1 + R / n) * (n * t)
    print(f"Amount (A) = {A}")

elif input1 == "AP":
    # Annuity Plan
    PMT = float(input("Enter Payment per period (PMT): "))
    R = float(input("Enter Rate (R in %): "))
    n = float(input("Enter number of periods per year (n): "))
    t = float(input("Enter time in years (t): "))
    A = PMT * ((1 + R / n) * (n * t) - 1)
    print(f"Amount (A) = {A}")

else:
    print("Invalid choice. Please run the program again.")
  
      
```
