```python
# Store girls data in separate lists
girls_names = ["Evelyn", "Jessica", "Somto", "Edith", "Liza", "Madonna", "Waje", "Tola", "Aisha", "Latifa"]
girls_ages = [17, 16, 17, 18, 16, 18, 17, 20, 19, 17]
girls_heights = [5.5, 6.0, 5.4, 5.9, 5.6, 5.5, 6.1, 6.0, 5.7, 5.5]
girls_scores = [80, 85, 70, 60, 76, 66, 87, 95, 50, 49]

# Store boys data in separate lists
boys_names = ["Chinedu", "Liam", "Wale", "Gbenga", "Abiola", "Kola", "Kunle", "George", "Thomas", "Wesley"]
boys_ages = [19, 16, 18, 17, 20, 19, 16, 18, 17, 19]
boys_heights = [5.7, 5.9, 5.8, 6.1, 5.9, 5.5, 6.1, 5.4, 5.8, 5.7]
boys_scores = [74, 87, 75, 68, 66, 78, 87, 98, 54, 60]

# Combine all data into one list
all_names = girls_names + boys_names
all_ages = girls_ages + boys_ages
all_heights = girls_heights + boys_heights
all_scores = girls_scores + boys_scores

print("Name\t\t| Age\t| Height\t| Score")
print("-" * 50)

# Print all students data
for i in range(20):  # 20 students total
    
    name = all_names[i]
    age = all_ages[i]
    height = all_heights[i]
    score = all_scores[i]
    
    # Print each row
    print(f"{name}\t\t| {age}\t| {height}\t\t| {score}")
```

<h1> Algorithm Pseudocode to display Student data in Tabular Format </h1>
    
    BEGIN
        // STEP 1: Initialize the data
        CREATE girls_names LIST with [Evelyn, Jessica, Somto, Edith, Liza, Madonna, Waje, Tola, Aisha, Latifa]
        CREATE girls_ages LIST with [17, 16, 17, 18, 16, 18, 17, 20, 19, 17]
        CREATE girls_heights LIST with [5.5, 6.0, 5.4, 5.9, 5.6, 5.5, 6.1, 6.0, 5.7, 5.5]
        CREATE girls_scores LIST with [80, 85, 70, 60, 76, 66, 87, 95, 50, 49]
    
        CREATE boys_names LIST with [Chinedu, Liam, Wale, Gbenga, Abiola, Kola, Kunle, George, Thomas, Wesley]
        CREATE boys_ages LIST with [19, 16, 18, 17, 20, 19, 16, 18, 17, 19]
        CREATE boys_heights LIST with [5.7, 5.9, 5.8, 6.1, 5.9, 5.5, 6.1, 5.4, 5.8, 5.7]
        CREATE boys_scores LIST with [74, 87, 75, 68, 66, 78, 87, 98, 54, 60]
        
        // STEP 2: Combine all data
        CREATE all_names LIST = girls_names + boys_names
        CREATE all_ages LIST = girls_ages + boys_ages
        CREATE all_heights LIST = girls_heights + boys_heights
        CREATE all_scores LIST = girls_scores + boys_scores
        
        // STEP 3: Print header for combined table
        PRINT "Name        | Age | Height | Score"
        PRINT "----------------------------------"
        
        // STEP 4: Loop through all students (0 to 19)
        FOR counter FROM 0 TO 19 DO
            // Get data for current student
            current_name = all_names[counter]
            current_age = all_ages[counter]
            current_height = all_heights[counter]
            current_score = all_scores[counter]
        
            // Print the row
            PRINT current_name + "        | " + current_age + "   | " + current_height + "      | " + current_score
        END FOR
        
    END
