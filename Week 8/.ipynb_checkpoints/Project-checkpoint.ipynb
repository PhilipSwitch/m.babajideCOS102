```python
import tkinter as tk
from tkinter import ttk, messagebox

# Parent Class
class Zenith:
    def __init__(self, division):
        self.division = division
    
    def unique_services(self):
        """To be overridden by child classes"""
        pass
    
    def mutual_services(self):
        """Common services across all divisions"""
        return ["Customer support", "Account statements", "Online banking", "Mobile banking"]
    
    def get_all_services(self):
        """Polymorphism: combines mutual and unique services"""
        mutual = self.mutual_services()
        unique = self.unique_services()
        return mutual + unique

# Child Class: Retail Banking
class RetailBanking(Zenith):
    def __init__(self):
        super().__init__("Retail Banking")
    
    def unique_services(self):
        """Overriding parent method - Polymorphism"""
        return [
            "Lines of credit",
            "Investment management and accounts",
            "Insurance",
            "Retirement and education accounts",
            "Loans and mortgages",
            "Checking and saving"
        ]

# Child Class: Global Banking
class GlobalBanking(Zenith):
    def __init__(self):
        super().__init__("Global Banking")
    
    def unique_services(self):
        """Overriding parent method - Polymorphism"""
        return [
            "Multi-currency management services and products",
            "Foreign currency accounts",
            "Foreign currency credit cards",
            "Transborder advisory services",
            "Liquidity management"
        ]

# Child Class: Commercial Banking
class CommercialBanking(Zenith):
    def __init__(self):
        super().__init__("Commercial Banking")
    
    def unique_services(self):
        """Overriding parent method - Polymorphism"""
        return [
            "Lines of credit",
            "Investment management and accounts",
            "Insurance",
            "Advisory services"
        ]

# GUI Application
class BankingApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Zenith Bank - Division Services")
        self.root.geometry("600x500")
        self.root.configure(bg='#f0f0f0')
        
        # Employee database
        self.employees = {
            "Mary": "Retail Banking",
            "Agatha": "Global Banking",
            "Noel": "Commercial Banking"
        }
        
        self.create_widgets()
    
    def create_widgets(self):
        # Title
        title_label = tk.Label(self.root, text="Zenith Bank Service Directory", 
                               font=("Arial", 18, "bold"), bg='#f0f0f0', fg='#004080')
        title_label.pack(pady=20)
        
        # Main Frame
        main_frame = tk.Frame(self.root, bg='#f0f0f0')
        main_frame.pack(pady=20)
        
        # Employee Selection Frame
        select_frame = tk.LabelFrame(main_frame, text="Employee Information", 
                                     font=("Arial", 12, "bold"), bg='#f0f0f0', padx=20, pady=10)
        select_frame.pack(pady=10, padx=20, fill="both")
        
        # Radio buttons for employee selection
        tk.Label(select_frame, text="Select Employee:", font=("Arial", 11), bg='#f0f0f0').grid(row=0, column=0, pady=10, sticky='w')
        
        self.employee_var = tk.StringVar()
        
        tk.Radiobutton(select_frame, text="Mary (Retail Banking)", variable=self.employee_var, 
                      value="Mary", bg='#f0f0f0', font=("Arial", 10)).grid(row=1, column=0, sticky='w', padx=20)
        tk.Radiobutton(select_frame, text="Agatha (Global Banking)", variable=self.employee_var, 
                      value="Agatha", bg='#f0f0f0', font=("Arial", 10)).grid(row=2, column=0, sticky='w', padx=20)
        tk.Radiobutton(select_frame, text="Noel (Commercial Banking)", variable=self.employee_var, 
                      value="Noel", bg='#f0f0f0', font=("Arial", 10)).grid(row=3, column=0, sticky='w', padx=20)
        
        # Or manual entry
        tk.Label(select_frame, text="Or Enter Manually:", font=("Arial", 11), bg='#f0f0f0').grid(row=4, column=0, pady=10, sticky='w')
        
        tk.Label(select_frame, text="Employee Name:", bg='#f0f0f0').grid(row=5, column=0, sticky='w')
        self.name_entry = tk.Entry(select_frame, width=30)
        self.name_entry.grid(row=5, column=1, pady=5, padx=10)
        
        tk.Label(select_frame, text="Division:", bg='#f0f0f0').grid(row=6, column=0, sticky='w')
        self.division_combo = ttk.Combobox(select_frame, values=["Retail Banking", "Global Banking", "Commercial Banking"], width=27)
        self.division_combo.grid(row=6, column=1, pady=5, padx=10)
        
        # Buttons Frame
        button_frame = tk.Frame(main_frame, bg='#f0f0f0')
        button_frame.pack(pady=20)
        
        tk.Button(button_frame, text="Get Services", command=self.get_services, 
                 bg='#004080', fg='white', font=("Arial", 11, "bold"), padx=20, pady=5).pack(side='left', padx=10)
        tk.Button(button_frame, text="Clear", command=self.clear_all, 
                 bg='#808080', fg='white', font=("Arial", 11), padx=20, pady=5).pack(side='left', padx=10)
        
        # Results Frame
        result_frame = tk.LabelFrame(main_frame, text="Services Rendered", 
                                     font=("Arial", 12, "bold"), bg='#f0f0f0', padx=20, pady=10)
        result_frame.pack(pady=10, padx=20, fill="both", expand=True)
        
        self.result_text = tk.Text(result_frame, width=60, height=15, font=("Arial", 10), wrap=tk.WORD)
        self.result_text.pack(pady=10)
        
        # Scrollbar for result text
        scrollbar = tk.Scrollbar(self.result_text)
        scrollbar.pack(side=tk.RIGHT, fill=tk.Y)
        self.result_text.config(yscrollcommand=scrollbar.set)
        scrollbar.config(command=self.result_text.yview)
    
    def get_services(self):
        """Get services based on selected or entered employee"""
        self.result_text.delete(1.0, tk.END)
        
        # Get employee info
        employee_name = self.employee_var.get()
        if not employee_name:
            employee_name = self.name_entry.get().strip()
        
        if not employee_name:
            messagebox.showwarning("Input Error", "Please select or enter an employee name!")
            return
        
        # Determine division
        division = None
        if employee_name in self.employees:
            division = self.employees[employee_name]
        else:
            division = self.division_combo.get()
            if not division:
                messagebox.showwarning("Input Error", "Please enter division for manual entry!")
                return
        
        # Create appropriate object (Polymorphism)
        if division == "Retail Banking":
            bank_obj = RetailBanking()
        elif division == "Global Banking":
            bank_obj = GlobalBanking()
        elif division == "Commercial Banking":
            bank_obj = CommercialBanking()
        else:
            messagebox.showerror("Error", f"Unknown division: {division}")
            return
        
        # Display results
        self.result_text.insert(tk.END, f"{'='*50}\n")
        self.result_text.insert(tk.END, f"EMPLOYEE: {employee_name.upper()}\n")
        self.result_text.insert(tk.END, f"DIVISION: {division.upper()}\n")
        self.result_text.insert(tk.END, f"{'='*50}\n\n")
        
        self.result_text.insert(tk.END, "MUTUAL SERVICES (All Divisions):\n")
        self.result_text.insert(tk.END, "-" * 35 + "\n")
        for service in bank_obj.mutual_services():
            self.result_text.insert(tk.END, f"• {service}\n")
        
        self.result_text.insert(tk.END, f"\nUNIQUE SERVICES ({division.upper()}):\n")
        self.result_text.insert(tk.END, "-" * 35 + "\n")
        for service in bank_obj.unique_services():
            self.result_text.insert(tk.END, f"• {service}\n")
        
        self.result_text.insert(tk.END, f"\nTOTAL SERVICES PROVIDED:\n")
        self.result_text.insert(tk.END, "-" * 35 + "\n")
        all_services = bank_obj.get_all_services()
        for i, service in enumerate(all_services, 1):
            self.result_text.insert(tk.END, f"{i}. {service}\n")
    
    def clear_all(self):
        """Clear all inputs and results"""
        self.employee_var.set("")
        self.name_entry.delete(0, tk.END)
        self.division_combo.set("")
        self.result_text.delete(1.0, tk.END)

# Main execution
if __name__ == "__main__":
    root = tk.Tk()
    app = BankingApp(root)
    root.mainloop()
```
